How is _online authenticity_ different from being authentic face-to-face?

### ASSISTANT
Online authenticity differs from face-to-face authenticity in several key ways:

Key Differences:

1. Control & Curation
- Online: You can carefully craft and edit your presentation
- Face-to-face: More spontaneous, harder to control reactions/expressions

2. Time Element
- Online: Can pause, reflect, and compose responses
- Face-to-face: Immediate reactions required, more "raw" interactions

3. Context
- Online: Often lacks nuance, body language, tone
- Face-to-face: Full sensory experience with non-verbal cues

4. Audience
- Online: Often speaking to multiple audiences simultaneously
- Face-to-face: Usually interacting with specific individuals/groups

5. Permanence
- Online: Content often remains accessible indefinitely
- Face-to-face: Interactions are typically temporary/ephemeral

This creates unique challenges online:
- Need to balance authenticity with professional image
- Managing different versions of self across platforms
- Navigating context collapse (when different social circles intersect)
